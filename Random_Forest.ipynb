```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.tree import DecisionTreeClassifier
```

```python
data=sns.load_dataset("titanic")
data.isnull().sum()
feature=["pclass","sex","fare","embarked","age"]
target=["survived"]
from sklearn.impute import SimpleImputer
#Fill the Null Value
num_imp=SimpleImputer(strategy='mean')
data[["age"]]=num_imp.fit_transform(data[["age"]])

cat_imp=SimpleImputer(strategy='most_frequent')
data[["embarked"]]=cat_imp.fit_transform(data[["embarked"]])
le=LabelEncoder()
data["sex"]=le.fit_transform(data["sex"])
data["embarked"]=le.fit_transform(data["embarked"])
X=data[feature]
Y=data["survived"]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
model=DecisionTreeClassifier()
model.fit(X_train,Y_train)
Y_predict=model.predict(X_test)
y_pred_train=model.predict(X_train)
from sklearn.metrics import accuracy_score
print(f"Accurecy of the Train Model is {accuracy_score(Y_train,y_pred_train)}")
print(f"Accurecy of the Model is {accuracy_score(Y_test,Y_predict)}")
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(
    n_estimators=201,
    oob_score=True
)
rf.fit(X_train,Y_train)
y_pred=rf.predict(X_test)
print(f"oob {rf.oob_score_ *100}")
print(f"Accurecy of the Model is {accuracy_score(Y_test,Y_predict)*100}")
    
```

```python
#bagging
from sklearn.ensemble import BaggingClassifier
base_model=DecisionTreeClassifier()
bagg=BaggingClassifier(
    base_model,
    n_estimators=201,
)
bagg.fit(X_train,Y_train)
Y_prd=bagg.predict(X_test)
print(f"Accurecy of the Model is {accuracy_score(Y_test,Y_prd)*100}")
```

```python
#bagging Logostic Regresion
from sklearn.linear_model import LogisticRegression
base_model=LogisticRegression(max_iter=100)
bagg=BaggingClassifier(
    base_model,
    n_estimators=201,
)
bagg.fit(X_train,Y_train)
Y_prd=bagg.predict(X_test)
print(f"Accurecy of the Model is {accuracy_score(Y_test,Y_prd)*100}")
```
