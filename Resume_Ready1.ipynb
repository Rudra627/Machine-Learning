```python
import pandas as pd
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
```

```python
data=pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
# data.info()
num_col=data.select_dtypes(include=["float64","int64"])
cat_col=data.select_dtypes(include=["object"])
data=data.drop(columns=["customerID"],axis=1)
data.head()
```

```python
#Encoding Part

lab_col=["gender","Partner","Dependents","PaperlessBilling","Churn","PhoneService"]
one_col = [
"InternetService",
"TechSupport",
"StreamingTV",
"StreamingMovies",
"Contract",
"PaymentMethod",
"MultipleLines",
"OnlineSecurity",
"OnlineBackup",
"DeviceProtection"
]


le=LabelEncoder()
for el in lab_col:
    data[el]=le.fit_transform(data[el])
# ohe=OneHotEncoder()
# for el in one_col:
#     data[[el]]=ohe.fit_transform(data[[el]])
data = pd.get_dummies(data, columns=one_col)
```

```python
bool_cols = data.select_dtypes(include='bool').columns
data[bool_cols] = data[bool_cols].astype(int)
```

```python
X=data.drop("Churn",axis=1)
Y= data["Churn"]
```

```python
X=X.drop(columns=["MonthlyCharges","TotalCharges"],axis=1)
```

```python
# from sklearn.preprocessing import StandardScaler

# scaler = StandardScaler()
# X_scaled = scaler.fit_transform(X)
X_train,X_test,Y_train,Y_test=train_test_split(
    X,Y,test_size=0.2,random_state=42)
```

```python
#Model
model=SVC()
model.fit(X_train,Y_train)
```

```python
Y_pred=model.predict(X_test)
print(f"Accuracy :{accuracy_score(Y_test,Y_pred)}")
```

```python
#Loggistic Regression
modelL=LogisticRegression()
modelL.fit(X_train,Y_train)
Y_pred=modelL.predict(X_test)
print(f"Accuracy :{accuracy_score(Y_test,Y_pred)}")
print("c",classification_report(Y_test,Y_pred))
```

```python
#Naive Bayece
modelN=GaussianNB()
modelN.fit(X_train,Y_train)
Y_pred=modelN.predict(X_test)
print(f"Accuracy :{accuracy_score(Y_test,Y_pred)}")
print("c",classification_report(Y_test,Y_pred))
```

```python
#Decission Tree
modelD=DecisionTreeClassifier()
modelD.fit(X_train,Y_train)
Y_pred=modelD.predict(X_test)
print(f"Accuracy :{accuracy_score(Y_test,Y_pred)}")
print("c",classification_report(Y_test,Y_pred))
```
