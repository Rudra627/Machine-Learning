```python
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.preprocessing import StandardScaler,LabelEncoder,OneHotEncoder
```

```python
data=pd.read_csv("customer_churn_dataset-training-master.csv")
data.info()
```

```python
from sklearn.impute import SimpleImputer
numeric_cols = data.select_dtypes(include=["float64", "int64"]).columns
categorical_cols = data.select_dtypes(include=["object"]).columns
num_imputer = SimpleImputer(strategy='mean')
data[numeric_cols] = num_imputer.fit_transform(data[numeric_cols])

# Impute categorical columns with most frequent
cat_imputer = SimpleImputer(strategy='most_frequent')
data[categorical_cols] = cat_imputer.fit_transform(data[categorical_cols])
```

```python
scale=StandardScaler()
# Exclude 'Churn' from numeric columns for scaling, as it is the target variable
numeric_cols_for_scaling = [col for col in numeric_cols if col != 'Churn']
data[numeric_cols_for_scaling]=scale.fit_transform(data[numeric_cols_for_scaling])
```

```python
# Encoding
data=data.drop("CustomerID",axis=1)
```

```python
label=LabelEncoder()
for el in categorical_cols:
  data[el]=label.fit_transform(data[el])
```

```python
x=data.drop("Churn",axis=1)
y=data["Churn"].astype(int) # Ensure the target variable is integer type for classification
x_train,x_test,y_train,y_test=train_test_split(
    x,y,test_size=0.2,random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.ensemble import (
    AdaBoostClassifier,
    BaggingClassifier,
    GradientBoostingClassifier,
    RandomForestClassifier,
    VotingClassifier,
    StackingClassifier,
    HistGradientBoostingClassifier, # For large datasets
    ExtraTreesClassifier
)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
#ensamble Learning

#randomForest
model1=RandomForestClassifier(n_estimators=500,criterion='gini')
model1.fit(x_train,y_train)
y_pred=model1.predict(x_test)
print("accuracy :",accuracy_score(y_test,y_pred))
```

```python

model2=AdaBoostClassifier(
    n_estimators=100, random_state=42
)
model2.fit(x_train,y_train)
```

```python
estimators = [
    ('rf', RandomForestClassifier(n_estimators=10, random_state=42)),
    ('svr', SVC()),
    ('log',LogisticRegression())
]
model3=StackingClassifier(
    estimators=estimators, final_estimator=LogisticRegression()
)
y_pred=model3.predict(x_test)

print("accuracy :",accuracy_score(y_test,y_pred))
```
